# ***SOLUCION TAREA MECANICA CUANTICA PUNTO D "Santiago Yepes *`Gaona`*"***

```python
import numpy as np
import matplotlib.pyplot as plt

c = 3e8

K_b = 1.381e-23

h = 6.62e-34

def I(la,T, c, K_b, h):

  return (2*h*c**2)/((la**5)*(np.exp((h*c)/(la*K_b*T))-1))


longitud_onda = np.logspace(-9, -5, 1000)


longitud_onda_nm = np.logspace(-9,-3,1000)



b = 2.898e-3

lamb_100 = b/100
lamb_5800 = b/5800
lamb_120000 = b/120000


plt.figure(1)
plt.plot(longitud_onda_nm, I(longitud_onda_nm, 100, c, K_b, h))
plt.axvline(x=lamb_100, color='r', linestyle='--', label=f'λmax para T=100ºK ')
plt.xlabel('Longitud de onda (nm)')
plt.ylabel('Intensidad (W/m^3)')
plt.title('Intensidad vs Longitud de onda Para 100º K')


plt.figure(2)
plt.plot(longitud_onda, I(longitud_onda, 5800, c, K_b, h))
plt.axvline(x=lamb_5800, color='b', linestyle='--',  label=f'λmax para T=5800ºK')
plt.xlabel('Longitud de onda (m)')
plt.ylabel('Intensidad (W/m^3)')
plt.title('Intensidad vs Longitud de onda Para 5800º K')

plt.figure(3)
plt.plot(longitud_onda, I(longitud_onda, 120000, c, K_b, h))
plt.axvline(x=lamb_120000, color='g', linestyle='--' ,label=f'λmax para T=120000ºK')
plt.xlabel('Longitud de onda (m)')
plt.ylabel('Intensidad (W/m^3)')
plt.title('Intensidad vs Longitud de onda Para 120000º K')

plt.show()


```
